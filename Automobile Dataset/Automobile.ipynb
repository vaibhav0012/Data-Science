# Automobile Dataset

Importing Libraries

```
import numpy as np
import pandas as pd
```

Import Dataset

```
data = pd.read_csv('Automobile price data _Raw_.csv')
```

Data Cleaning


Replace ? string with null values

```
data.replace('?', np.nan ,inplace = True)
```

Check null values

```
data.isnull().sum()
```

Drop values with high missing data count


```
data = data.drop('normalized-losses',axis=1)
```

Check data types

```
data.dtypes
```

Data Transformation

Defining columns for transformation

```
col_int = ['horsepower','peak-rpm','price']
col_float = ['bore','stroke']
columns = ['bore','stroke','horsepower','peak-rpm','price']
```

Check for values in num-of-doors column


```
data['num-of-doors'].value_counts()
```

Fill maximum occuring value inplace of NULL values


```
data['num-of-doors'].replace(np.nan,'four',inplace =True)
```

Transforming categorical columns to numerical

```
str_to_num = {"four": 4, "six": 6, "five": 5, "eight": 8,"two": 2, "twelve": 12, "three":3 }
data.replace(str_to_num, inplace=True)
str_to_num_cat = {"convertible":5,"hatchback":1,"sedan":4,"wagon":3,"hardtop":2}
data.replace(str_to_num_cat, inplace=True)
```

Transforming Numerical Data

```
data[columns] = data[columns].fillna(0)
data[col_int] = data[col_int].astype(int)
data[col_float] = data[col_float].astype(float)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=0, strategy='mean')
imputer.fit(data[columns])
data[columns] = imputer.transform(data[columns])
```

Label Encoder for categorical values

```
Cat_col = ['make','fuel-type','aspiration','drive-wheels','engine-location','engine-type','fuel-system']
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
data[Cat_col] = data[Cat_col].apply(lb.fit_transform)
```

Setting input and output parameters

```
Y = data['price'].values.reshape(-1,1)
X = data.drop('price',axis=1)
```

Standardizing the values

```
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_Y = StandardScaler()
X[columns[0:-1]] = sc_X.fit_transform(X[columns[:-1]])
Y = sc_Y.fit_transform(Y)
```

Reducing dimensionality for a better output

```
from sklearn.decomposition import PCA
pca = PCA(n_components = 22)
X = pca.fit_transform(X)
```

Spliting the dataset

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

Kernal SVM Model

```
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, Y_train)
```

Random Forest Method

```
from sklearn.ensemble import RandomForestRegressor
regressor1 = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor1.fit(X_train, Y_train)
```

Predicting the model

```
Y_pred = regressor.predict(X_test)
Y_pred1 = regressor1.predict(X_test)
```

Model Accuracy check

```
from sklearn.metrics import mean_squared_error, r2_score 
print("SVM")
print('MSE: ',mean_squared_error(Y_test,Y_pred)) 
print(r2_score(Y_test,Y_pred))
print("Random Forest")
print('MSE: ',mean_squared_error(Y_test,Y_pred1))
print(r2_score(Y_test,Y_pred1))
```
