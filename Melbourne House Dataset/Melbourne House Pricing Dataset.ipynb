### Dropping Duplicates

```python
dataset.shape
```

```python
dataset = dataset.drop_duplicates(['Suburb','Address','Rooms','SellerG','Date'],keep = 'first',inplace = False)
```

#### We can see that some people made the same booking for the same room on the same data that shows replicas in the dataseta nd hence were removed from the dataset.

#### The postCode is given as a float data type but it can be represented as a string

```python
dataset['Postcode'] = dataset['Postcode'].astype(str)
```

```python
dataset.dtypes
```

### Check for Null Values

```python
dataset.isnull().sum()
```

#### The Columns Region Name,PostCode  and Council area are categorical values so we fill the maximum occuring data in null values

```python
dataset['CouncilArea'].replace(np.nan,'Boroondara City Council',inplace = True)
dataset['Regionname'].replace(np.nan,'Southern Metropolitan',inplace = True)
dataset['Postcode'].replace(np.nan,'3073',inplace = True)
```

```python
dataset.isnull().sum()
```

```python
col_num_flt = ['Price','Distance','Landsize','BuildingArea','Lattitude','Longtitude']
col_num_int = ['Bedroom2','Bathroom','Car','YearBuilt']
```

#### we can see that for different missing values of Propert count we have multiple missing values such as Distance, Bedroom. Bedroom etc which cannot be gathere from existing suburb columns hence we'll drop the missing rows in case of PropertCount

```python
dataset.dropna(subset=['Propertycount'],axis=0,inplace=True)
```

#### Imputing Float values data

```python
from sklearn.impute import SimpleImputer
imputer_flt = SimpleImputer(missing_values=np.nan, strategy='mean')
dataset[col_num_flt] = imputer_flt.fit_transform(dataset[col_num_flt])
```

#### Imputing Integer values

```python
dataset[col_num_int] = imputer_flt.fit_transform(dataset[col_num_int])
dataset[col_num_int]=dataset[col_num_int].astype(int)
```

```python
dataset[col_num_int]
```

#### Searching for outliners

```python
dataset.describe(include = 'all')
```

```python
plt.figure(figsize=(16,18))
plt.subplot(3,2,1)
sns.distplot(dataset['Price'])
plt.title('Price')
plt.subplot(3,2,2)
sns.distplot(dataset['Distance'])
plt.title('Distance')
plt.subplot(3,2,3)
sns.distplot(dataset['Rooms'])
plt.title('Rooms')
plt.subplot(3,2,4)
sns.distplot(dataset['Bathroom'])
plt.title('Bathrooms')
plt.subplot(3,2,5)
sns.distplot(dataset['Car'])
plt.title('Car')
plt.subplot(3,2,6)
sns.distplot(dataset['Landsize'])
plt.title('Land Size')
```

#### We observe there are many outliners in case of LandSize so we need to drop out the outliners to get a better outpur

```python
q = dataset['Landsize'].quantile(0.98)
dataset = dataset[dataset['Landsize']<q]
q = dataset['Price'].quantile(0.97)
dataset = dataset[dataset['Price']<q]
q = dataset['Rooms'].quantile(0.99)
dataset = dataset[dataset['Rooms']<q]
q = dataset['Bathroom'].quantile(0.99)
dataset = dataset[dataset['Bathroom']<q]
q = dataset['Car'].quantile(0.99)
dataset = dataset[dataset['Car']<q]
q = dataset['BuildingArea'].quantile(0.99)
dataset = dataset[dataset['BuildingArea']<q]
```

```python
dataset.shape
```

```python
plt.figure(figsize=(16,18))
plt.subplot(3,2,1)
sns.distplot(dataset['Price'])
plt.title('Price')
plt.subplot(3,2,2)
sns.distplot(dataset['Distance'])
plt.title('Distance')
plt.subplot(3,2,3)
sns.distplot(dataset['Rooms'])
plt.title('Rooms')
plt.subplot(3,2,4)
sns.distplot(dataset['Bathroom'])
plt.title('Bathrooms')
plt.subplot(3,2,5)
sns.distplot(dataset['Car'])
plt.title('Car')
plt.subplot(3,2,6)
sns.distplot(dataset['Landsize'])
plt.title('Land Size')
```

#### We can see that now the data is normalised

#### Standardizing the data

```python
col_std=['Price','Distance','Landsize','BuildingArea','Bedroom2','Bathroom','Car','YearBuilt','Propertycount']
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
dataset[col_std] = sc.fit_transform(dataset[col_std])
```

```python
dataset = dataset.drop(['Lattitude','Longtitude','Address','Method','Date'],axis=1)
```
